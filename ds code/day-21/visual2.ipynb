```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
df.plot(kind='scatter', x='age', y='fare', alpha=0.5, color = 'red')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.show()                                                                          
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('Fare')
plt.xlabel('gender')
plt.show()  
```

```python
# plt plot for age and fare
plt.scatter(df['age'], df['fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
# seaborn plot for age and fare
sns.scatterplot(data=df, x='age', y='fare')
plt.show()
```

```python
#seaborn plot more types
sns.pairplot(df)
plt.show()
```

```python
sns.barplot(data=df,x='survived',y='age')
plt.show()
```

```python
#fare of groupby gender and survived
df.groupby(['sex', 'survived'])['fare'].mean().plot(kind='bar')
```

```python
sns.barplot(data=df,x='sex',y='fare', hue='survived')
```

```python
sns.boxplot(data=df,x='sex',y='fare', hue='survived')
```

```python
sns.scatterplot(data=df,x='age',y='fare', hue='sex')
```
