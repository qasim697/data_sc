### **1. Categorical Data (Comparing Categories)**
- **Bar Chart**: Best for comparing quantities across different categories (e.g., sales by product, population by country).
- **Stacked Bar Chart**: Useful for showing the composition of categories within a larger group (e.g., sales by product and region).
- **Pie Chart**: Ideal for showing proportions or percentages of a whole (e.g., market share by company). Use sparingly and only when you have a small number of categories.


```python
import pandas as pd
import matplotlib.pyplot as plt

months = pd.date_range(start='2023-01-01', periods=12, freq='M')
sales = [100, 120, 130, 150, 200, 250, 300, 280, 230, 200, 180, 160]

# Line Chart
plt.plot(months, sales, marker='o', color='green')
plt.title('Monthly Sales in 2023')
plt.xlabel('Month')
plt.ylabel('Sales (in $1000)')
plt.grid(True)
plt.show()
```
