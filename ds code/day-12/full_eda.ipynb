```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
# Step 1: Initial Data Inspection
print("Step 1: Initial Data Inspection")
print("--------------------------------")
print("First 5 rows of the dataset:")
print(df.head())
print("\nShape of the dataset:", df.shape)
print("\nData types of each column:")
print(df.info())
print("\nMissing values in each column:")
print(df.isnull().sum())
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python
# Step 2: Data Cleaning
print("\nStep 2: Data Cleaning")
print("----------------------")
# Handle missing values
df['Age'].fillna(df['Age'].median(), inplace=True)  # Fill missing age with median
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)  # Fill missing Embarked with mode
df.drop('Cabin', axis=1, inplace=True)  # Drop Cabin column (too many missing values)

# Check if missing values are handled
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# Step 3: Univariate Analysis
print("\nStep 3: Univariate Analysis")
print("----------------------------")
# Histogram for Age
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.savefig('age_distribution.png')  # Save the plot
plt.show()
```

```python
# Bar plot for Survived
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count')
plt.savefig('survival_count.png')  # Save the plot
plt.show()
```

```python
# Bar plot for Pclass
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', data=df, palette='Set3')
plt.title('Passenger Class Distribution')
plt.savefig('pclass_distribution.png')  # Save the plot
plt.show()
```

```python
# Step 4: Bivariate Analysis
print("\nStep 4: Bivariate Analysis")
print("---------------------------")
# Survival rate by Gender
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived', data=df, palette='Set1')
plt.title('Survival Rate by Gender')
plt.savefig('survival_rate_by_gender.png')  # Save the plot
plt.show()
```

```python
# Survival rate by Passenger Class
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Survived', data=df, palette='Set2')
plt.title('Survival Rate by Passenger Class')
plt.savefig('survival_rate_by_pclass.png')  # Save the plot
plt.show()
```

```python
# Scatter plot for Age vs. Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='Set1')
plt.title('Age vs. Fare')
plt.savefig('age_vs_fare.png')  # Save the plot
plt.show()
```

```python
# Step 5: Multivariate Analysis
print("\nStep 5: Multivariate Analysis")
print("------------------------------")
# Heatmap for correlation matrix
plt.figure(figsize=(8, 6))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('correlation_matrix.png')  # Save the plot
plt.show()
```

```python
# Pair plot for numerical variables
sns.pairplot(df[['Age', 'Fare', 'Survived']], hue='Survived', palette='Set1')
plt.savefig('pair_plot.png')  # Save the plot
plt.show()
```

```python
 #Step 6: Feature Engineering
print("\nStep 6: Feature Engineering")
print("----------------------------")
# Create a new feature: Family Size
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
# Create a new feature: Age Group
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 18, 35, 50, 100], labels=['Child', 'Young Adult', 'Adult', 'Elderly'])
```

```python
# Display the new features
print("\nNew features added:")
print(df[['FamilySize', 'AgeGroup']].head())
```

```python
# Step 7: Outlier Detection
print("\nStep 7: Outlier Detection")
print("--------------------------")
# Boxplot for Fare
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Fare'])
plt.title('Boxplot for Fare')
plt.savefig('fare_boxplot.png')  # Save the plot
plt.show()
```

```python
# Handle outliers in Fare
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Fare'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]
```

```python
# Step 8: Insights and Conclusions
print("\nStep 8: Insights and Conclusions")
print("---------------------------------")
print("Key Insights:")
print("- Females and higher-class passengers had a higher survival rate.")
print("- Most passengers were between 20 and 40 years old.")
print("- Higher fares were associated with higher survival rates.")
print("- Passengers with smaller families had a better chance of survival.")
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_titanic.csv', index=False)
print("\nCleaned dataset saved as 'cleaned_titanic.csv'.")
```
